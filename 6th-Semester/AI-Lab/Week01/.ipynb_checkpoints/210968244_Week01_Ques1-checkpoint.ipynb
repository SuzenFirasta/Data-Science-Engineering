# Random Policy Agent in OpenAI Gym Environment

This script demonstrates a basic random policy agent in an OpenAI Gym environment. The agent takes random actions at each step and visualizes its performance over a specified number of steps.

## Setup

- `num_steps`: Number of steps the agent runs for.
- `env`: The OpenAI Gym environment.

## Initialization

- `obs`: Reset the environment to its initial state.
- `cum_reward`: Cumulative reward for the current episode.
- `rewards`: List to store cumulative rewards for each episode.
- `steps`: List to store the number of steps for each episode.

## Main Loop

The script iterates through the specified number of steps, taking random actions at each step.

- At each step, the agent:
  - Takes a random action from the environment's action space.
  - Applies the action to the environment and receives feedback (observation, reward, termination status, etc.).
  - Renders the environment for visualization.
  - Accumulates the reward for the current step.

- If the episode is terminated:
  - Records the cumulative reward and the number of steps for the episode.
  - Resets the environment for the next episode.

## Visualization

After running the agent, the script closes the environment and plots the recorded steps versus cumulative rewards to visualize the agent's performance.

Note: Adjust the time delay (`time.sleep(0.1)`) based on preferences for frame visualization speed.


```python
# Number of steps you run the agent for 
num_steps = 1000

obs = env.reset()

cum_reward = 0
rewards = []
steps = []

for step in range(num_steps):
    # take random action
    action = env.action_space.sample()
    
    # apply the action
    obs, reward, terminated, truncated, info = env.step(action)
    cum_reward+=reward
    
    # Render the env
    env.render()

    # Wait a bit before the next frame unless you want to see a crazy fast video
    time.sleep(0.1)
    
    # If the epsiode is up, then start another one
    if (terminated):
        rewards.append(cum_reward)
        steps.append(step)
        cum_reward = 0
        env.reset()

# Close the env
env.close()

plt.plot(steps, rewards)
```
