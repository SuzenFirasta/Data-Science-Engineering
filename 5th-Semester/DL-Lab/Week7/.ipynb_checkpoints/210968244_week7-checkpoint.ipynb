removing URLs

```python
import re
# Define a function to remove URLs using a regular expression
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# Apply the remove_urls function to the 'text' column
df['text'] = df['text'].apply(remove_urls)

# Print the DataFrame with URLs removed
print(df)
```

punctuation Removal

```python

import string
df['text'] = df['text'].apply(lambda x: ''.join([char for char in x if char not in string.punctuation]))
# Print the DataFrame with URLs removed
print(df)
```

```python
import nltk
from nltk.corpus import stopwords
# Download the NLTK stop words list (if not already downloaded)
nltk.download('stopwords')

# Create a set of English stop words from NLTK
stop_words = set(stopwords.words('english'))

# Define a function to remove stop words from a text
def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# Apply the remove_stopwords function to the 'text' column
df['text'] = df['text'].apply(remove_stopwords)

# Print the DataFrame with stop words removed
print(df)
```

```python
import pandas as pd
from textblob import TextBlob
from tqdm import tqdm

# Define a function to correct spelling in a text
def correct_spelling(text):
    blob = TextBlob(text)
    return str(blob.correct())

# Apply the correct_spelling function to the 'text' column with tqdm for progress
tqdm.pandas(desc="Correcting Spelling")
df['text'] = df['text'].progress_apply(correct_spelling)

# Print the DataFrame with spelling corrected
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk import FreqDist, bigrams
from nltk.tokenize import word_tokenize

# Combine the 'text' column into a single string
all_text = " ".join(df['text'])

# Tokenize the combined text into words
words = word_tokenize(all_text)

# Calculate the frequency distribution of words
word_freq = FreqDist(words)

# Get the top N most frequent words and their counts
top_words = word_freq.most_common(30)  # Change 30 to the desired number of words to display

# Extract words and counts from the top_words list
word_labels, word_counts = zip(*top_words)

# Create a bar plot for the most frequent words using Matplotlib
plt.figure(figsize=(12, 6))
plt.barh(word_labels, word_counts, color='skyblue')
plt.title("Top Most Frequent Words")
plt.xlabel("Frequency")
plt.ylabel("Words")
plt.gca().invert_yaxis()  # Invert the y-axis to display the most frequent words at the top
plt.show()

# Create bigrams from the words
bi_grams = list(bigrams(words))

# Calculate the frequency distribution of bigrams
bigram_freq = FreqDist(bi_grams)

# Get the top N most frequent bigrams and their counts
top_bigrams = bigram_freq.most_common(30)  # Change 30 to the desired number of bigrams to display

# Extract bigrams and counts from the top_bigrams list
bigram_labels, bigram_counts = zip(*top_bigrams)

# Convert bigram tuples to strings for labeling
bigram_labels = [" ".join(bigram) for bigram in bigram_labels]

# Create a bar plot for the most frequent bigrams using Matplotlib
plt.figure(figsize=(12, 6))
plt.barh(bigram_labels, bigram_counts, color='lightcoral')
plt.title("Top Most Frequent Bigrams")
plt.xlabel("Frequency")
plt.ylabel("Bigrams")
plt.gca().invert_yaxis()  # Invert the y-axis to display the most frequent bigrams at the top
plt.show()
```

```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

def prerposLemmaToken(df, max_words):
    # Lowercase text
    df['text'] = df['text'].str.lower()
    
    # Remove URLs
    df['text'] = df['text'].apply(lambda text: re.sub(r'https?://\S+|www\.\S+', '', text))
    
    # Remove punctuation, brackets, and symbols
    df['text'] = df['text'].apply(lambda text: re.sub(r'[^\w\s]', '', text))
    
    # Remove <br> and new lines
    df['text'] = df['text'].replace('<br>', ' ', regex=True)
    df['text'] = df['text'].replace('\n', ' ', regex=True)
    
    # Tokenize
    df['text'] = df['text'].apply(lambda text: ' '.join(word_tokenize(text)))
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    df['text'] = df['text'].apply(lambda text: ' '.join([word for word in word_tokenize(text) if word not in stop_words]))
    
    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    df['text'] = df['text'].apply(lambda text: ' '.join([lemmatizer.lemmatize(word) for word in word_tokenize(text)]))
    
    # Create a tokenizer and fit it on the text data
    tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
    tokenizer.fit_on_texts(df['text'])
    
    # Convert text data to numerical sequences
    sequences = tokenizer.texts_to_sequences(df['text'])
    
    # Pad sequences to a fixed length
    max_sequence_length = max(len(seq) for seq in sequences)
    padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length, padding='post', truncating='post')
    
    # Get sequences and labels
    sequences = padded_sequences
    labels = df['label'].tolist()
    sequences = np.array(sequences)
    labels = np.array(labels)
    return sequences, labels
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
import numpy as np  # Import numpy

def lstm_model(sequences, labels, max_words=10000, num_epochs=10, batch_size=32, validation_split=0.2):
    # Convert sequences and labels to numpy arrays
    sequences = np.array(sequences)
    labels = np.array(labels)

    # Split the data into training and validation sets
    train_sequences, valid_sequences, train_labels, valid_labels = train_test_split(
        sequences, labels, test_size=validation_split, random_state=42
    )

    # Define your model architecture
    max_sequence_length = sequences.shape[1]  # Get the sequence length from the input data

    model_lstm = Sequential()
    model_lstm.add(Embedding(input_dim=max_words, output_dim=128, input_length=max_sequence_length))
    for _ in range(15):
        model_lstm.add(LSTM(64, return_sequences=True))
    model_lstm.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model and get the training history
    history = model_lstm.fit(train_sequences, train_labels, epochs=num_epochs, batch_size=batch_size, validation_data=(valid_sequences, valid_labels))

    return model_lstm, history
```
